```python
#Imports
# Used to organize the dataset into a table
import pandas as pd
# Used for creating charts and graphs
import matplotlib.pyplot as plt
# Loads the built-in Iris dataset
from sklearn.datasets import load_iris
# Standardizes the data before clustering
from sklearn.preprocessing import StandardScaler
# Imports the clustering algorithms
from sklearn.cluster import KMeans, DBSCAN
# Used to compare our clusters to the actual flower species
from sklearn.metrics import adjusted_rand_score
# Loads the Iris dataset from scikit-learn
iris = load_iris()

# Creates a pandas DataFrame so the data is easier to work with
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

# Adds the numerical target values to the DataFrame
df["target"] = iris.target

# Converst the numerical target into flower names
df["species"] = df["target"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})

# Displays the first five rows of the dataset
print("First Five Rows")
print(df.head())

# Displays summary statistics for each feature
print("\nSummary Statistics")
print(df.describe())

# Checks if there are any missing values
print("\nMissing Values")
print(df.isnull().sum())

# Part B - K-Means Clustering

# Selects only the feature columns
X = df[iris.feature_names]

# Creates a StandardScaler object
# Scaling helps K-Means because it uses distance calculations.
scaler = StandardScaler()

# Scales all feature values
X_scaled = scaler.fit_transform(X)

# Different K values I want to test
k_values = [2, 3, 4, 5, 6]

# Dictionary to save the clustering results
kmeans_results = {}

# Loops through each K value
for k in k_values:

    # Creates a K-Means model
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    # Trains the model and assign every flower to a cluster
    clusters = kmeans.fit_predict(X_scaled)

    # Saves the clusters
    kmeans_results[k] = clusters

    # Compares our clusters to the actual Iris species
    ari = adjusted_rand_score(df["target"], clusters)

    # Prints the clustering score
    print(f"K = {k}")
    print(f"Adjusted Rand Index = {ari:.4f}")

# Part C - DBSCAN Clustering
# Try several combinations of eps and min_samples
dbscan_settings = [
    {"eps": 0.3, "min_samples": 5},
    {"eps": 0.5, "min_samples": 5},
    {"eps": 0.7, "min_samples": 5},
    {"eps": 0.5, "min_samples": 3}
]

# Loops through each parameter combination
for setting in dbscan_settings:

    # Saves the values so they are easier to read
    eps_value = setting["eps"]
    min_samples_value = setting["min_samples"]

    # Creates the DBSCAN model
    dbscan = DBSCAN(
        eps=eps_value,
        min_samples=min_samples_value
    )

    # Trains the model and predict clusters
    dbscan_clusters = dbscan.fit_predict(X_scaled)

    # Counts the number of clusters found
    number_of_clusters = len(set(dbscan_clusters)) - (
        1 if -1 in dbscan_clusters else 0
    )

    # Counts the number of noise points
    number_of_noise_points = list(dbscan_clusters).count(-1)

    # Compares DBSCAN to the actual species
    ari = adjusted_rand_score(
        df["target"],
        dbscan_clusters
    )

    # Displays the results
    print(f"eps = {eps_value}")
    print()
    print(f"min_samples = {min_samples_value}")
    print()
    print(f"Clusters = {number_of_clusters}")
    print()
    print(f"Noise Points = {number_of_noise_points}")
    print()
    print(f"ARI = {ari:.4f}")
```
